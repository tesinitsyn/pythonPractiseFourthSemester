Создайте генератор случайных спрайтов 5x5 пикселей. Используйте свойство симметрии. Вывод реализуйте с помощью функции imshow.

Пример работы генератора показан ниже.

![](data/sprite1.png)

````python
import numpy as np
import matplotlib.pyplot as plt


def generate_sprite():
    # создаем случайный массив 5x5
    sprite = np.random.randint(0, 2, size=(5, 5))
    # применяем свойство симметрии
    sprite = np.maximum(sprite, sprite[::-1])
    sprite = np.maximum(sprite, sprite.T[::-1].T)
    return sprite


# создаем несколько случайных спрайтов
sprites = [generate_sprite() for _ in range(5)]

# выводим спрайты с помощью функции imshow
fig, axs = plt.subplots(1, len(sprites), figsize=(10, 2))
for ax, sprite in zip(axs, sprites):
    ax.imshow(sprite, cmap='gray')
    ax.axis('off')
plt.show()
````

**2.2.** (уровень сложности: средний)

Создайте генератор карт спрайтов. Пример работы этого генератора показан ниже.


![](data/sprite2.png)

````python
import numpy as np
import matplotlib.pyplot as plt


def generate_sprite():
    # создаем случайный массив 5x5
    sprite = np.random.randint(0, 2, size=(5, 5))
    # применяем свойство симметрии
    sprite = np.maximum(sprite, sprite[::-1])
    sprite = np.maximum(sprite, sprite.T[::-1].T)
    return sprite


def generate_sprite_map(n, m):
    # создаем карту спрайтов размером n x m
    sprite_map = np.zeros((n * 5, m * 5))
    for i in range(n):
        for j in range(m):
            # вычисляем координаты спрайта в карте спрайтов
            x = i * 5
            y = j * 5
            # генерируем спрайт и добавляем его в карту
            sprite = generate_sprite()
            sprite_map[x:x + 5, y:y + 5] = sprite
    return sprite_map


# создаем карту спрайтов размером 6x6
sprite_map = generate_sprite_map(6, 6)

# выводимкарту спрайтов в формате, аналогичном примеру изображения
fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(1 - sprite_map, cmap='gray', vmin=0, vmax=1, interpolation='nearest')
ax.set_aspect('equal')
ax.axis('off')
plt.show()
````

````python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

# Размер спрайта NxM
N, M = 64, 64

# Создание пустого спрайта
sprite = np.zeros((N, M, 3), dtype=np.uint8)

# Определение случайных центров симметрии
centers = [(random.randint(0, N - 1), random.randint(0, M - 1)) for _ in
           range(3)]

# Определение случайных осей симметрии
axes = [random.randint(0, 1) for _ in range(3)]

# Определение случайных цветов
colors = [np.array(
    [random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)])
          for _ in range(3)]

# Заполнение спрайта с помощью свойств симметрии и цветов
for i in range(N):
    for j in range(M):
        for k in range(3):
            if axes[k] == 0:
                # Ось симметрии - вертикаль
                sprite[i, j, k] = colors[k][
                    (i - centers[k][0]) % len(colors[k])]
            else:
                # Ось симметрии - горизонталь
                sprite[i, j, k] = colors[k][
                    (j - centers[k][1]) % len(colors[k])]

# Отображение спрайта
norm = Normalize(vmin=0, vmax=255)
plt.imshow(sprite, norm=norm, interpolation='nearest')
plt.show()

````

Создайте генератор растений с помощью алгоритма освоения пространства (space colonization algorithm). Этот алгоритм описан в [статье](http://algorithmicbotany.org/papers/colonization.egwnp2007.large.pdf) 2007 года и позволяет моделировать различные природные явления: корни и ветви деревьев, сети рек и дорог и так далее.

Основная идея алгоритма состоит в том, чтобы разместить в некоторой области случайным образом точки притяжения. К этим точкам на каждой итерации тянутся узлы моделируемого растения или дерева. В процессе создаются ветки, а точки поглощаются.

Алгоритм содержит следующие шаги.

1. Создание N случайных точек притяжения, а также одного или нескольких начальных узлов дерева.
2. Поглощение ближайших точек притяжения – то есть тех точек, которые находятся к узлам дерева ближе, чем расстояние $d_k$.
3. Нахождение для каждой точки притяжения ближайшего к ней узла дерева $v$ на дистанции, не превышающей $d_i$. Для каждого $v$ формируется множество $S(v)$, которое может содержать несколько точек притяжения.
4. Если $S(v)$ не пусто, то формируется очередной узел $v'$ на расстоянии $D$ и создается новый сегмент $(v, v')$ с помощью следующих формул:
$$
v' = v + \hat{n}D, \hat{n} = \frac{\vec{n}}{\lVert\vec{n}\rVert},\vec{n} = \sum_{s \in S(v)} \frac{s - v}{\lVert s - v \rVert}
$$

**2.7.** (уровень сложности: низкий)

Реализуйте функцию, которая по имени графического файла возвращает двумерный массив с помощью списков Питона, содержащий RGB-значения картинки. Эта функция понадобится для решения остальных задач этого раздела.

````python
from PIL import Image


def read_image(filename):
    with Image.open(filename) as img:
        # преобразуем изображение в RGB-формат
        img = img.convert("RGB")
        # получаем размеры изображения
        width, height = img.size
        # создаем двумерный массив для хранения RGB-значений каждого пикселя
        pixels = [[0] * width for _ in range(height)]
        # проходимся по каждому пикселю в изображении и сохраняем его RGB-значение в массиве
        for y in range(height):
            for x in range(width):
                pixels[y][x] = img.getpixel((x, y))
        return pixels


image_pixels = read_image("image.png")
print(image_pixels)
````

Диаграммы Вороного находят широкое применение не только в математике, но и, к примеру, в архитектуре, а также в процедурной графике. Названы эти диаграммы в честь русского математика [Георгия Феодосьевича Вороного](https://www.mathnet.ru/links/bd2cf5586728b0e7a7393c5d7cc7b23d/cheb698.pdf). Для создания диаграммы Вороного необходимо выбрать множество точек и разбить плоскость (n-мерный случай рассматривать не будем) на области, для каждой из которых только одна из выбранных точек окажется ближайшей.

Мы используем диаграммы Вороного для реализации фильтра изображения. Результат фильтрации может напоминать нарисованную кистью картину или мозаику. Для получения результата необходимо выбрать N случайных точек исходного изображения и построить по ним диаграмму. 

Существуют различные алгоритмы построения диаграмм Вороного, но мы воспользуемся самым примитивным подходом – просто последуем определению и переберем все координаты изображения-результата на предмет расстояния до случайно выбранных точек исходного изображения.

````python
import random
from PIL import Image
from math import sqrt


def euclidean_distance(a, b):
    """
    Функция, которая вычисляет расстояние между двумя точками a и b
    на плоскости по формуле Евклида.
    """
    return sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)


def create_voronoi_diagram(source_image, num_points):
    """
    Функция, которая создает диаграмму Вороного на основе указанного изображения
    и заданного количества случайно выбранных точек.
    """
    # Сгенерировать случайные точки на изображении
    image_width, image_height = source_image.size
    points = [(random.randint(0, image_width - 1),
               random.randint(0, image_height - 1)) for _ in range(num_points)]

    # Создать двумерный массив, чтобы хранить цвета каждой области диаграммы
    diagram = [[(0, 0, 0) for _ in range(image_width)] for _ in
               range(image_height)]

    # Проходимся по всем координатам изображения
    for x in range(image_width):
        for y in range(image_height):
            # Находим ближайшую точку на изображении
            distances = [euclidean_distance((x, y), point) for point in points]
            min_distance = min(distances)
            min_index = distances.index(min_distance)

            # Записываем цвет области в соответствующую ячейку массива
            diagram[y][x] = source_image.getpixel(points[min_index])

    return diagram


def read_image(filename):
    with Image.open(filename) as img:
        # преобразуем изображение в RGB-формат
        img = img.convert("RGB")
        # получаем размеры изображения
        width, height = img.size
        # создаем двумерный массив для хранения RGB-значений каждого пикселя
        pixels = [[0] * width for _ in range(height)]
        # проходимся по каждому пикселю в изображении и сохраняем его RGB-значение в массиве
        for y in range(height):
            for x in range(width):
                pixels[y][x] = img.getpixel((x, y))

        # Создаем диаграмму Вороного на основе массива RGB-значений
        voronoi_diagram = create_voronoi_diagram(img,
                                                 50)  # измените значение 50 на нужное количество случайных точек

        # Обновляем массив RGB-значений на основе диаграммы Вороного
        for y in range(height):
            for x in range(width):
                pixels[y][x] = voronoi_diagram[y][x]

        return pixels


image_pixels = read_image("image.png")
print(image_pixels)
````

Реализуйте простой [алгоритм](https://github.com/s-macke/VoxelSpace) трехмерной визуализации воксельных ландшафтов на основе текстуры и карты высот.

Исходные данные: [текстура](data/texture.png), [карта высот](data/heightmap.png).

Идея алгоритма заключается в послойном сканировании, одна горизонтальная линия за другой, значений из текстуры с учетом дистанции от камеры, и размещении на экране каждой точки слоя в виде столбца, высота которого извлекается из карты высот, а цвет – из текстуры.


````python
import pygame
from PIL import Image

# Загрузка текстуры и карты высот
texture = Image.open("texture.png").convert("RGBA")
heightmap = Image.open("heightmap.png").convert("L")
width, height = heightmap.size

# Создание окна с помощью библиотеки Pygame
pygame.init()
screen = pygame.display.set_mode((width, height))
clock = pygame.time.Clock()

# Вычисление максимальной высоты из карты высот
max_height = float(max(heightmap.getdata()))

# Основной цикл программы
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()

    # Очистка экрана и установка начальных значений
    screen.fill((0, 0, 0))
    y_offset = height - 1

    # Проход по каждой строке карты высот и текстуры
    for y in range(height):
        for x in range(width):
            # Нахождение высоты точки на текущей позиции
            height_value = heightmap.getpixel((x, y))
            height_value_normalized = height_value / max_height
            distance = min(1.0, float(height_value_normalized) * 4.0)

            # Нахождение цвета текстуры на текущей позиции
            texture_color = texture.getpixel((x, y))

            # Вычисление цвета точки на экране
            color = (int(texture_color[0] * distance),
                     int(texture_color[1] * distance),
                     int(texture_color[2] * distance))

            # Размещение точки на экране в виде столбца
            pygame.draw.rect(screen, color, (x, y_offset, 1, -height_value))

        y_offset -= 1

    # Обновление экрана
    pygame.display.update()

    # Ограничение количества кадров в секунду
    clock.tick(60)

````
